```python
#타이타닉 or data.go.kr의 데이터로 오늘 학습한 내용들을 연습

##001 - object 타입으로 저장된 날짜 데이터를 datetime 형식으로 바꾸기 연습
air=pd.read_csv("data/Seoul.csv")

air['dt_date']=pd.to_datetime(air['측정일시'], format='%Y%m%d%H')
air
```

```python
##002 - 위에 datetime으로 만든 dt_date열을 인덱스로 만들기
air.index=air['dt_date']
air
```

```python
##003 - dt_date에서 연도만 추출해서 dt_year라는 새로운 열을 만들자
air['dt_year']=air['dt_date'].dt.year
air
```

```python
##005 - Tesla 데이터로 연습하기
tesla=pd.read_csv("data/TSLA.csv")
tesla['dt_date']=pd.to_datetime(tesla['Date'])
tesla['year']=tesla['dt_date'].dt.year
year=tesla.groupby('year')['Close'].mean()

fig,ax=plt.subplots()
ax=year.plot()
plt.show()
```
