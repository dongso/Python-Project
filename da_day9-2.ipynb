```python
import numpy as np
#정렬 : sort, axix=0(행),1(열)
a=np.array([[4,3,2,5],[1,12,21,6],[9,7,8,25]])
np.sort(a)
np.sort(a,axis=1)
np.sort(a,axis=-1) #axis=-1 가장 안쪽의 차원
```

```python
#기술통계 : 사분위수, 중앙값, 최대/최소, 표준편차...
x=np.array([5,1,3,2,4,8,7,50]) #1,2,3,4,5,7,8,50
print(len(x)) #8
print(np.mean(x))#10.0
print(np.var(x)) #분산 : 233.5
print(np.std(x)) #표준편차 : 15.2807...
print(np.max(x)) #최대값 : 50
print(np.median(x)) #중앙값 :4.5 => 데이터가 짝수면 중앙에 있는 값 두개의 평균을 반환.
print(np.percentile(x,0)) #사분위 값( data, n) n%에 해당하는 값을 반환함.
print(np.percentile(x,25))
print(np.percentile(x,50))
print(np.percentile(x,75))
print(np.percentile(x,100))
```

```python
#데이터 개수 세기
x=np.array([10,10,1,1,20,20,70])
x2=np.unique(x) #중복된 값을 제거하고, unique한 값만 출력됨.
print(x2)

res1, res2=np.unique(x, return_counts=True) #unique한 값들만 출력한 후에 그 값이 몇개 존재하는지도 같이 반환.
print(res1)
print(res2)
```

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
#sklean의 패지지의 dataset모듈에 있는 load_digits함수를 가져와라

digits=load_digits()
print(digits['data'])
samples=[0,10,20,30,1,11,21,31]
d=[]

for i in range(8):
    d.append(digits.images[samples[i]])
    
plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1,8,i+1)
    plt.imshow(d[i],cmap=plt.cm.bone_r,interpolation='nearest') #숫자 데이터를 이미지로 변환하여 출력
    
    plt.title("img{}".format(i+1))
    
    
plt.tight_layout()    
plt.show()
```

```python

v=[]
print(np.shape(d))
for i in range(8):
    v.append(d[i].reshape(64,1)) #8 X 8 => 64 X 1
    
plt.figure(figsize=(8,3))
for i in range(8):
    plt.subplot(1,8,i+1)
    plt.imshow(v[i],cmap=plt.cm.bone_r,interpolation='nearest') #숫자 데이터를 이미지로 변환하여 출력
    
    plt.title("img{}".format(i+1))
    
    
plt.tight_layout(w_pad=11)    
plt.show()
```
