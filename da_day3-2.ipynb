    #시각화 과정
    1) 전체 그래프가 출력될 기본 틀(figure) 생성
    2) 틀을 그래프 격자로 나눈다.
    3) 각 격자에 그래프를 추가한다.
    4) 각 격자에 그래프 출력


```python
#1) 전체 그래프가 출력될 기본 틀(figure) 생성
fig=plt.figure()

#2) 틀을 그래프 격자로 나눈다.
axes1=fig.add_subplot(2,2,1) #2x2 로 나눈다음의 첫번째 위치를 axes1로 저장.
axes2=fig.add_subplot(2,2,2)
axes3=fig.add_subplot(2,2,3)
axes4=fig.add_subplot(2,2,4)

#3) 각 격자에 그래프를 추가한다.

axes1.plot(dataset_1['x'],dataset_1['y'],'p') #point 옵션을 p,s,o등으로 주면 됨
axes2.plot(dataset_2['x'],dataset_2['y'],'s')
axes3.plot(dataset_3['x'],dataset_3['y'],'o')
axes4.plot(dataset_4['x'],dataset_4['y'],'o')

#4) 각 격자에 그래프 출력
axes1.set_title("dataset_1")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")

fig.suptitle("Anscombe Data")
fig.tight_layout()
```

```python
#시각화 도구 : matplotlib, seaborn

tips=sns.load_dataset("tips")
tips
```

```python
#축 : axis(단수), axes(복수)
fig=plt.figure()
axes1=fig.add_subplot(1,1,1) #1x1 에서 1번째
axes1.set_xlabel("Total_Bill")
axes1.set_ylabel("Frequency")
axes1.set_title("Histogram of Total Bill")
#axes1.hist(tips['total_bill'])
axes1.hist(tips['total_bill'],bins=20) #구간이 20개로 나눠짐
```

```python
#산점도 그래프 : 변수 2개로 그린 그래프, 이변량 그래프라고도 함.
#식사 비용에 따른 팁 열의 분포를 산점도 그래프로 나타내보자.

scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(x=tips['total_bill'],y=tips['tip']) #(x축 데이터, y축 데이터)
axes1.set_xlabel("Total_Bill")
axes1.set_ylabel("Tip")
axes1.set_title("Scatterplot of Total Bill vs Tip")
```

#Boxplot => 이산형 변수와 연속형 변수를 함께 사용하는 그래프 & outliers를 파악하기 좋다
#성별에 따른 팁 횟수

이산형 변수(명확하게 구분되는 값), 연속형 변수(명확하게 셀 수 없는 범위의 값)

```python
box_plot=plt.figure()
axes1=box_plot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex']=='Female']['tip'],
              tips[tips['sex']=='Male']['tip']],
             labels=['Female','Male'])
#boxplot([ 첫번째 출력될 데이터    ,  두번째 출력될 데이터,   ...  ],labels[첫번째 데이터의 label, 두번째 데이터의 label,  ...])
```

```python
#다변량 그래프 그리기
def recode_sex(sex):
    if sex=="Female":
        return 0
    else: return 1
```

```python
tips['sex']
#apply(): 특정 데이터에 대해 특정 함수를 적용하고자 할 때 사용
# 데이터.apply(적용할 함수명)

tips['sex_color']=tips['sex'].apply(recode_sex)
tips
```

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)

axes1.set_xlabel("Total_Bill")
axes1.set_ylabel("Tip")
axes1.set_title("Total Bill vs Tip - 3D")

axes1.scatter(x=tips['total_bill'],
              y=tips['tip'],
              c=tips['sex_color'], #값마다 색상 지정 (여기서는 0과 1)
              alpha=0.8 #alpha -> 투명도
             )
```

```python
df3=pd.DataFrame({
    'a':[1,2,3,4],
    'b':[2,5,6,7],
    'c':[6,1,2,6]
})
df3
```

```python
df3['a'].max()
df3['a'].min()
```

```python
df3.apply(lambda x: x.max()) #각 열의 최대값 찾기
#df3의 각 열이 lambda의 x로 사용됨.
#따라서, 위 식의 의미는 각 열의 최대값을 추출해내는 것이다.

df3.apply(lambda x: x.max()-x.min())
#위 식의 의미, 각 열의 최대값-최소값

df3.apply(lambda x: x.max()-x.min(),axis=1) #axis=1을 줘서 행 단위로 람다 함수가 실행되게 함
```
