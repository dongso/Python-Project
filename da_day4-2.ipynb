```python
import matplotlib as mpl
import matplotlib.pylab as plt
%matplotlib inline
import seaborn as sns

tips=sns.load_dataset("tips")
tips
```

```python
#seaborn 에서 히스토그램 그릴 때 : subplots(기본틀), distplot(데이터 전달)메서드를 사용하여 시각화
ax=plt.subplots()
#ax=sns.distplot(tips['total_bill'],kde=True, hist=True) #kde=False :: 밀집도 그래프 제외,hist=False:: 히스토그램 그래프를 제외
#히스토그램과 밀집도 그래프가 함께 출력(넓이가 1인 그래프)


ax=sns.distplot(tips['total_bill'],rug=True) #rug::양탄자 그래프-> 직선으로 데이터의 밀집 정도를 출력
```

```python
ax=plt.subplots()
ax=sns.countplot('day',data=tips) #day 컬럼의 데이터가 몇번씩 나타났는지를 확인
ax.set_title("Count of Days")
tips
```

```python
ax=sns.regplot(x='total_bill',y='tip', data=tips, fit_reg=False) #fit_reg=False:: 회귀선 출력 안함
#산점도 그래프 + 회귀선 => regplot()
```

```python
#jointplot()::산점도 그래프 + 히스토그램
ax=sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex') 
# 1-<=pearson 상관계수<=1 , 0에 가까울 수록 상관관계가 적다. 
```

```python
#kdeplot() :: 2차원 밀집도 그래프
ax=sns.kdeplot(data=tips['total_bill'],data2=tips['tip'],shade=True) 
```

```python
#시간(Lunch / Dinner)에 따라 지불한 비용의 평균을 바 그래프로 출력
ax=sns.barplot(x="time", y="total_bill", data=tips)
ax.set_title("title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
```

```python
#시간(Lunch / Dinner)에 따라 지불한 비용의 평균을 박스 그래프로 출력
ax=sns.boxplot(x="time", y="total_bill", data=tips)
ax.set_title("title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
```

```python
#시간(Lunch / Dinner)에 따라 지불한 비용의 평균을 바이올린 그래프로 출력
ax=sns.violinplot(x="time", y="total_bill", data=tips)
ax.set_title("title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
```

```python
#pariplot():: 관계 그래프 -> 조합 그래프
fig=sns.pairplot(tips)
```

```python
#여러개의 그래프를 PairGrid 연산을 이용해 대각선의 위, 아래, 대각선으로 지정해 줄 수 있다.
pair_grid=sns.PairGrid(tips)
pair_grid.map_upper(sns.regplot)
pair_grid.map_lower(sns.kdeplot)
pair_grid.map_diag(sns.distplot,rug=True)
```
