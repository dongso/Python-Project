```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
x = iris.petal_length.values #꽃잎의 길이
y= iris.petal_width.values
```

```python
sns.rugplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Rug Plot ")
plt.show()
sns.boxplot(data=x)
plt.title("꽃잎의 길이에 대한 boxPlot")
```

```python
sns.kdeplot(x,y, shade=True)
```

```python
sns.distplot(x, rug=True, kde=True)
sns.jointplot(x='sepal_length', y='petal_length',data=iris)
```
