    axes 생성방법
    1) subplot:figure 내부에 axes를 생성, 그리드 형태로 axes가 생성됨
        figure가 행렬이면 axes는 행렬의 요소!
    2) plot(): subplot대신 plot만 사용해도 알아서 axes를 만들어냄
    ex. subplot(2,3,1) 2x3 행렬을 만든후 1번째 위치에 plot을 그리겠다는 명령

```python
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)
y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

ax1=plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')

ax2=plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
```

twinx함수 : x축은 공통으로 사용하면서, y축 값을 서로 다르게 표시할 때 사용

```python
fig, ax0=plt.subplots() #생략시 1개 axes가 default로 생성된다
ax1=ax0.twinx() #x축 공유, y축 개별 사용.
ax0.plot([10,5,2,9,7],'r--',label='y0')
ax1.plot([100,500,300,700,200],'b:', label='y1')
fig.legend(loc=0)
```

```python
#bar chart : x축: 카테고리
#bar() 세로막대 , barh() : 가로막대
y=[2,3,4]
x=np.arange(len(y))
xlabel=['가','나','다']
plt.title("바 차트")
plt.bar(x,y)
plt.xticks(x,xlabel)
plt.yticks(y, sorted(y)) #sorted(y)를 하지 않아도 내부적으로 sorting을 한 다음에 오름차순으로 정렬시켜 보여준다.
```

```python
people=['길동','꺽정','순신','세종']
y_pos=np.arange(len(people))
performance=3+10*np.random.rand(len(people)) #random number가 people의 수 만큼 나옴.(0-1사이)
error=np.random.rand(len(people))
plt.yticks(y_pos,people)
print(performance)
plt.barh(y_pos,performance,xerr=error, alpha=0.75) #y_pos의 값을 people에 저장된 값으로 치환해서 출력함.
#error bar 추가 , error가 있을 수 있는 정도를 표시해줄 때 사용함.
```

```python
#입력 변수가 2개, 출력변수가 1개 => 3차원 
#3차원 데이터를 시각화 => 등고선(contour)
#등고선 출력함수 : contour-순수 등고선, counterf-등고선+색

def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

x=np.linspace(-3,3,256) #1x256
y=np.linspace(-3,3,256)

xx,yy=np.meshgrid(x,y) #그리드 행렬로 만들어버림. 256x256
#meshgrid된 데이터를 가지고 등고선을 그릴 수 있다.
#xx,yy는 입력 데이터, zz는 출력데이터
zz=f(xx,yy)
plt.contourf(xx,yy,zz, cmap='jet', alpha=0.75)
plt.contour(xx,yy,zz,colors='black') #등고선의 선을 그어줌. contourf없이 단독으로 사용하면 선만 존재

# print(xx.shape)
# print("="*100)
# print(yy)
```

```python
x=[1,3,5] #1x3
#y=[2,4,6]
y=[2,4] #1x2
X,Y=np.meshgrid(x,y) #2x3 => x,y의 shape을 맞추기 위해서 meshgrid를 사용한다.
print(X)
print("="*50)
print(Y)
print("="*50)
Z=X+Y
print(Z)
print("="*50)
Z=X*Y #행렬의 곱셈은 아님, 같은자리에 위치한 요소간 곱셈이 이뤄짐.(elemenet-wise product)
#행렬의 곱셈 : [3,2]*[2,1]=[3,1]
print(Z)
print("="*50)
Z=np.sqrt(X**2+Y**2)
print(Z)

plt.contourf(X,Y,Z,10) #뒤에 10은 레벨을 나타낸다.
plt.contour(X,Y,Z,10,colors='black')
```
