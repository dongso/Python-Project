```python
import numpy as np
import pandas as pd
```

```python
# 1) pivot연습
data = pd.DataFrame({'cust_id': ['c1', 'c1', 'c1', 'c2', 'c2', 'c2', 'c3', 'c3', 'c3'], 
'prod_cd': ['p1', 'p2', 'p3', 'p1', 'p2', 'p3', 'p1', 'p2', 'p3'],
'grade' : ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B'], 
'pch_amt': [30, 10, 0, 40, 15, 30, 0, 0, 10]})
```

```python
data
```

```python
data_pivot=data.pivot(index='cust_id', columns='prod_cd', values='pch_amt')
data_pivot
```

```python
#pivot_table(data, index, columns, values) 을 주로 사용한다.
data_pivotTab=pd.pivot_table(data,index='cust_id', columns='prod_cd', values='pch_amt')
data_pivotTab
```

```python
#data_pivot=data.pivot(index=['cust_id','grade'], columns='prod_cd', values='pch_amt') 
#index를 두개 지정할 수 없다.

pd.pivot_table(data,index=['cust_id','grade'], columns='prod_cd', values='pch_amt')
#index를 두개 지정할 수 있다.

#--> 보통 pivot_table을 더 많이 사용한다.

#컬럼이 2개인 경우
#pivot -> 지정 불가
#data.pivot(index='cust_id', columns=['prod_cd','grade'], values='pch_amt')

#pivot_table
pd.pivot_table(data,index='cust_id', columns=['grade','prod_cd'], values='pch_amt')
```

```python
data
```

```python
#data.pivot(index='grade',columns='prod_cd',values='pch_amt') #=> Error::중복된 값이 있어서 value Error가 발생됨
pd.pivot_table(data, index='grade',columns='prod_cd',values='pch_amt') #default로 평균값이 적용된다.
```

```python
pd.pivot_table(data, index='grade',columns='prod_cd',values='pch_amt',aggfunc=np.count_nonzero)
#default로 평균값이 적용된다.
#aggfunc 옵션에 사용할 함수를 기입해주면 그 함수를 사용해 집계를 하게된다.
```

```python
#pd.pivot_table(data, index='grade',columns='prod_cd',values='pch_amt',aggfunc=np.sum, margins=True) #margin을 사용하면 행과 열의 합계+모든 값의 합계를 추가해줌
pd.pivot_table(data, index='grade',columns='prod_cd',values='pch_amt',aggfunc=np.mean, margins=True) 
#margin을 사용하면 aggfunc에 지정한 함수에 따라 All값을 해당함수로 적용시킨후 데이터를 출력해준다.
```

```python
ebola=pd.read_csv("data/country_timeseries.csv")
ebola.head()
ebola.columns
```

```python
#"Deaths_Mali".split("_")
variable_split=ebola_long.variable.str.split("_")
variable_split[0][1]
```

```python
status_values=variable_split.str.get(0)
country_values=variable_split.str.get(1)
status_values[-5:]
ebola_long.head()
ebola_long['status']=status_values
ebola_long['country']=country_values
ebola_long.head()
```

```python
#여러개의 열을 하나로 정리하기~!
weather=pd.read_csv("data/weather.csv")
weather.head()
#날짜 열을 행 데이터로 피봇

weather_melt=pd.melt(weather, id_vars=['id','year','month','element'],var_name='day', value_name='temp')
weather_melt

#연습하기
#weather_melt를 변경해보자. id, year, month, day가 열 인덱스, 컬럼 인덱스 element, value는 temp로...
weather_tidy=pd.pivot_table(weather_melt, index=['id','year','month','day'], columns='element', values='temp')
weather_tidy 


weather_tidy_flat=weather_tidy.reset_index()
weather_tidy_flat.head()
```

```python
billboard_long=pd.melt(billboard, id_vars=['year','artist','track','time','date.entered'], var_name='Week', value_name='Rating')
billboard_long.shape
billboard_long.info()

#track이 Loser인 곡 찾기
billboard_long[billboard_long['track']=='Loser']

billboard_songs=billboard_long[['year','artist','track','time']]
print(billboard_songs.shape)
#위의 데이터는 중복이 많음 -> 중복을 제거해 보자!!


billboard_songs=billboard_songs.drop_duplicates()
print(billboard_songs.shape)
len(billboard_songs)
billboard_songs['id']=range(len(billboard_songs))
billboard_songs
```
